```python
image3 = cv2.imread(path)
print(type(image3))
  
# Shape of the image
print("Shape of the image", image3.shape)
  
# [rows, columns]
crop = image3[50:180, 100:300]  
  
cv2.imshow('original', image3)
cv2.imshow('cropped', crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
image4 = cv2.imread(path, 1)
# Loading the image
 
half = cv2.resize(image4, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image4, (1050, 1610))
 
stretch_near = cv2.resize(image4, (780, 540),
               interpolation = cv2.INTER_NEAREST)
 
 
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image4, half, bigger, stretch_near]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```

```python
# Python program to explain cv2.rotate() method
  
# importing cv2
import cv2
# Reading an image in default mode
src = cv2.imread(path)
  
# Using cv2.rotate() method
# Using cv2.ROTATE_90_CLOCKWISE rotate
# by 90 degrees clockwise
image5 = cv2.rotate(src, cv2.ROTATE_90_CLOCKWISE)
  
# Displaying the image
cv2.imshow('Image rotated', image5)
cv2.waitKey(0)
```

```python
# Reading an image in default mode
image6 = cv2.imread(path)
# Declaring matrix
matrix = np.ones(image6.shape, dtype = "uint8") * 120
image6 = cv2.add(image6, matrix)
# Displaying the image
cv2.imshow('Image brightened', image6)
cv2.waitKey(0)
```

```python
# Reading an image in default mode
image7 = cv2.imread(path)
#MedianBlur
image7 = cv2.medianBlur(image7,5)
# Displaying the image
cv2.imshow('Median Blur', image7)
cv2.waitKey(0)
```

```python
# Reading an image in default mode
image8 = cv2.imread(path, 1)
# Sharpening the image
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharpened  = cv2.filter2D(src=image8, ddepth=-1, kernel=kernel)
# Displaying the image
cv2.imshow('Bird Sharpened', sharpened)
cv2.waitKey()
cv2.destroyAllWindows()
```
